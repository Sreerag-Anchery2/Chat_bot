```python
import nltk
from nltk.stem import WordNetLemmatizer
lemmatiser=WordNetLemmatizer()

words=[]
classes=[]
documents=[]
ignore_words=['?','!']
```

```python
for intent in intents['intents']:
    for vals in intent['patterns']:
#         print(vals)
        w=nltk.word_tokenize(vals)
        words.extend(w)
        documents.append((w,intent['tag']))
        if intent['tag'] not in classes:
            classes.append(intent['tag'])
            
print(words)
print(documents)
print(classes)
print(len(words))
print(len(documents))
print(len(classes))
```

```python
training=[]
output_empty=[0]*len(classes)
print(output_empty)
```

```python
documents
```

```python
for doc in documents:
    bag=[]
    patter_word=doc[0]
    patter_word=[lemmatiser.lemmatize(word.lower()) for word in patter_word]
    #print(patter_word)
    for w in words:
        bag.append(1) if w in patter_word else bag.append(0)
    output_row=list(output_empty)
    output_row[classes.index(doc[1])]=1
    training.append([bag,output_row])
print(training)
```

```python
x=[]
y=[]
```

```python
for i in training:
    x.append(i[0])
    y.append(i[1])
```

```python
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from keras.optimizers import SGD
import numpy as np
```

```python
model = Sequential()
model.add(Dense(128, input_shape=(len(x[0]),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(y[0]), activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
hist=model.fit(np.array(x),np.array(y),epochs=100,batch_size=5,verbose=1)
```

```python
import numpy as np
import random
import nltk
from nltk.stem import WordNetLemmatizer
lemmatise = WordNetLemmatizer()
```

```python
def clean_up_sentence(sentense):
    sentense_words=nltk.word_tokenize(sentense)
    sentense_words=[lemmatise.lemmatize(word.lower()) for word in sentense_words]
    return sentense_words

def bow(sentense,words,show_details=False):
    sentense_words=clean_up_sentence(sentense)
    bag=[0]*len(words)
    for s in sentense_words:
        for i,w in enumerate(words):
            if w==s:
                bag[i]=1
                if show_details:
                    print('found in bag')
    return np.array(bag)

def predict_class(sentense,model):
    p=bow(sentense,words,show_details=False)
    res=model.predict(np.array([p]))[0]
    ERROR_THRESHOLD=0.25
    results=[[i,r] for i,r in enumerate(res) if r>ERROR_THRESHOLD]
    results.sort(key=lambda x:x[1],reverse=True)
    return_list=[]
    for r in results:
        return_list.append({'intent':classes[r[0]],'probability':str(r[1])})
    return return_list
```

```python
def response(ints,intents_json):
    tags=ints[0]['intent']
    list_of_intent=intents_json['intents']
    for i in list_of_intent:
        if (i['tag']==tags):
            result=random.choice(i['responses'])
    return result
```

```python
def chat_bot_Response(msg):
    ints=predict_class(msg,model)
    res=response(ints,intents)
    return res
```

```python
import tkinter as tk
from tkinter import scrolledtext

def send_message():
    msg = message_entry.get().strip()
    if msg:
        chat_log.config(state=tk.NORMAL)
        chat_log.insert(tk.END, "You: " + msg + '\n')
        response = chat_bot_Response(msg)
        chat_log.insert(tk.END, "Bot: " + response + '\n\n')
        chat_log.config(state=tk.DISABLED)
        chat_log.yview(tk.END)
    message_entry.delete(0, tk.END)

def chat_bot_Response(msg):
    ints=predict_class(msg,model)
    res=response(ints,intents)
    return res

# Create the main window
root = tk.Tk()
root.title("Chat Application")
root.geometry("450x500")

# Create and configure chat log (ScrolledText widget)
chat_log = scrolledtext.ScrolledText(root, wrap=tk.WORD, state=tk.DISABLED, font=("Arial", 12))
chat_log.pack(padx=10, pady=10, fill=tk.BOTH, expand=True)

# Create a frame for entry and button
input_frame = tk.Frame(root)
input_frame.pack(fill=tk.X, padx=10, pady=10)

# Create and configure message entry
message_entry = tk.Entry(input_frame, font=("Arial", 12))
message_entry.pack(side=tk.LEFT, fill=tk.X, expand=True, padx=5)

# Create and configure send button
send_button = tk.Button(input_frame, text="Send", font=("Arial", 12), bg="#32de97", activebackground="#3c9d9b", fg="#ffffff", command=send_message)
send_button.pack(side=tk.RIGHT)

# Bind the Enter key to the send_message function
root.bind('<Return>', lambda event: send_message())

# Start the main event loop
root.mainloop()
```
